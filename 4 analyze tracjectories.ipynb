```python
conditionname='diabetes'
```

```python
#save results in CSV: encounter types
with open(path+conditionname+'_DISCO.csv', "w") as csv_file:        
    writer = csv.writer(csv_file, delimiter=',')
    combined_row=[]
    combined_row.append('pid')
    combined_row.append('date')                
    combined_row.append('type')
    writer.writerow(combined_row)
    for pid in data:   
        if len(data[pid]['appt'])>0:   
            for date in data[pid]['appt']:
                if data[pid]['appt'][date]['type']!='0':
                    combined_row=[]
                    combined_row.append(pid)
                    combined_row.append(date)       
                    combined_row.append(data[pid]['appt'][date]['type'])
                    writer.writerow(combined_row)
            if data[pid]['mortalitydate']!='NA':
                combined_row=[]
                combined_row.append(pid)
                combined_row.append(date)                
                combined_row.append('deceased')
                writer.writerow(combined_row)
```

```python
#save results in CSV: encounter details
with open(path+conditionname+'_DISCOVT.csv', "w") as csv_file:        
    writer = csv.writer(csv_file, delimiter=',')
    combined_row=[]
    combined_row.append('pid')
    combined_row.append('date')                
    combined_row.append('encounter')
    writer.writerow(combined_row)
    for pid in VTdict:   
        if len(VTdict[pid])>0:   
            for date in VTdict[pid]:
                    combined_row=[]
                    combined_row.append(pid)
                    combined_row.append(date)       
                    combined_row.append(VTdict[pid][date])
                    writer.writerow(combined_row)
            if data[pid]['mortalitydate']!='NA':
                combined_row=[]
                combined_row.append(pid)
                combined_row.append(date)                
                combined_row.append('deceased')
                writer.writerow(combined_row)
```
